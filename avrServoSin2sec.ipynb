```python
# Interactive Python Notebook
# Calculating a table of 8 bit signed values that will be added to a 16 bit pulse width 
# to oscillate a servo motor in a sinusoidal motion.

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# pwMax and pwMin determined by bench testing servo motor with signal generator
# desired travTime will be the approximate time the servo takes to move from full
#     left to full right.
# deltaT = 20 mS (Period for requred 50 Hz pulse to servo motor)
# steps is calculated number of steps required to cover travTime
# AVR program must load calculated pwMid to pulse width register and output to servo during setup

pwMax = 2350
pwMin = 450
pwRange = pwMax - pwMin
pwMid = int((pwMax + pwMin) / 2)
travTime = 2.0
deltaT = .02
steps = int(travTime / deltaT)

print("  Max:", pwMax, "  Min:", pwMin, "  Mid:", pwMid, "  Range:", pwRange, "  Steps:", steps)

# Generate list of sin() values with peaks at pwMax and pwMin
data = []
for t in range(steps):
    pos = int(pwRange / 2 * np.sin(t * 2 * np.pi / steps) + pwMid)
    data.append(pos)

print("\nPulse Widths:\n", data)
plt.plot(data)
plt.show()
print("\nCalculated Min:", min(data), "  Calculated Max:", max(data))

# Generate list of step sizes that can be added to Pulse Width register
stepSize = []
stepSizeSigned = []
for i, d in enumerate(data):
    size = data[(i + 1)%len(data)] - d
    stepSize.append(size)
    stepSizeSigned.append(size if size >= 0 else 256 + size)

print("\nStep Size\n", stepSize)

plt.plot(stepSize)
plt.show()

print("\nStep Size 8 bit Signed\n", stepSizeSigned)
print("\n\nExample adding 8 bit signed step size to 16 bit Pulse Width Register high and low bytes named pwH:pwL")
print("using TempH and TempL to hold step size value\n")

# Printing Assembly Language Function to read next value in Flash Memory
print("\
\n\tld  \tTempL, X+\t;X = SRAM address of signed step size value\
\n\tclr \tTempH\
\n\tsbrc\tTempL, 7\t;Check sign bit and set high byte to 0xFF if it is set\
\n\tser \tTempH\
\n\tadd \tpwL, TempL\n\tadc\tpwH, TempH\
\n")

# Generating a string containing all 8 bit signed hex values formatted as a table with .db (define byte)
# directives that can be included in the AVR program.

TaddLable = "TABLEaddr" + str(travTime).replace(".","_")
TendLable = "TABLEend" + str(travTime).replace(".","_")

Table = TaddLable + ":\n.db\t"
for i, s in enumerate(stepSizeSigned):
    Table += "$" + hex(s)[2:].zfill(2).upper()
    Table += "\n.db\t" if i % 8 == 7 and i < len(stepSizeSigned) - 1 else\
    "\n" if i >= len(stepSizeSigned) - 1 else ",\t"
Table += TendLable + ":"

print("\nOption using data table stored in FLASH Memory\
\n; Typically SRAM space is limited so it may be advantageous to keep\
\n; data tables in FLASH memory.  Special consideration must be taken\
\n; when addressing word addressed Flash memory\
\n; Z Register is used for the LPM instruction.  (Load Program Memory)\
\n; Set Z address to beginning of data table.\
\n; Program Memory uses WORD addressing.  2 Bytes for each address.\
\n; LPM uses BYTE addressing.  1 Byte for each address.\
\n; Multiply WORD address by 2 to calculate BYTE address.\
\n; Shifting a Byte Left once is the the same as multiplying it by 2.\
\n; (Value << 1) = (Value * 2)\n\
\n " + Table + "\n\
\n;directives requred before setup:\
\n.equ\tpwMid = ", pwMid, "\
\n.def\tTempH = R17\
\n.def\tTempL = R16\
\n.def\tpwH = R19\
\n.def\tpwL = R18\
\n\n;Instructions required during setup\
\n\tldi \tpwH, high(pwMid)\t;pwMid value = ", pwMid, "\
\n\tldi \tpwL, low(pwMid)\
\n\tldi \tZH, high(" + TaddLable + " << 1)\
\n\tldi \tZL, low(" + TaddLable + " << 1)\
\n")

# Printing Assembly Language Function to read next value in StepSize table
print("\
\n;Load and Store ZH:ZL as needed (It may be used by other code)\
\n;Function to read next value from table in Flash Memory  \
\nreadNextStep:\
\n\tlpm \tTempL, Z+\t;Z = FLASH Memory address of signed step size value\
\n\tclr \tTempH\
\n\tsbrc\tTempL, 7\t;Check sign bit and set high byte to 0xFF if it is set\
\n\tser \tTempH\
\n\tadd \tpwL, TempL\
\n\tadc \tpwH, TempH\
\n\tldi \tTempH, high(" + TendLable + " << 1)\
\n\tldi \tTempL, low(" + TendLable + " << 1)\
\n\tcp  \tZL, TempL\
\n\tcpc \tZH, TempH\
\n\tbrne\tnot" + "TABLEend" + str(travTime).replace('.', '_') + "\
\n\tldi \tZH, high(" + TaddLable + " << 1)\
\n\tldi \tZL, low(" + TaddLable + " << 1)\
\nnotTableEnd" + str(travTime).replace('.','_') + ":\
\n\tret\
")
```
