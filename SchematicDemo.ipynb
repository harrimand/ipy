```python
import numpy as np

#Resistor Voltages
R1 = 20
R2 = 30
R3 = 40
R4 = 20
R5 = 10
R6 = 50
R7 = 60
R8 = 30

#Loop Resistance Sums
RL1 = R1 + R2 + R3 + R7 + R8
RL2 = R2 + R3 + R4
RL3 = R4 + R5 + R6
RL4 = R5 + R7

#Loop Voltages
VL1 = 12
VL2 = 0
VL3 = 6
VL4 = 9

# Loop Resistor Matrix
LoopR = np.matrix([[-RL1,    R2+R3,  0,    R7],
                   [ R2+R3, -RL2,    R4,   0],
                   [ 0,      R4,    -RL3,  R5],
                   [ R7,     0,      R5,  -RL4]])
np.disp(LoopR)

print('\n')

# Loop Voltage Matrix
LoopV = np.matrix([[-VL1],[-VL2],[-VL3],[-VL4]])
np.disp(LoopV)

# Loop_Currents = Resistor_Matrix ^-1 * Voltage_Matrix
I = LoopR.I * LoopV

I1, I2, I3, I4 = I.tolist()  # I[0][0], I[1], I[2], I[3]

print('\nLoop Currents:')
print('I1 = ', *I1)
print('I2 = ', *I2)
print('I3 = ', *I3)
print('I4 = ', *I4)

# Sum voltage drops on each loop
VLoop1 = (I[0] * RL1 - I[1]*(R2 + R3) - I[3] * R7).tolist()[0][0]
VLoop2 = (-I[0] * (R2+R3) + I[1] * RL2 - I[2] * R4).tolist()[0][0]
VLoop3 = (-I[1] * R4 + I[2] * RL3 - I[3] * R5).tolist()[0][0]
VLoop4 = (-I[0] * R7 - I[2] * R5 + I[3] * RL4).tolist()[0][0]

print('\nResistor Voltage Sums:')
print('Loop 1: ', VLoop1)
print('Loop 2: ', VLoop2)
print('Loop 3: ', VLoop3)
print('Loop 4: ', VLoop4)

```
