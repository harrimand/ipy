```python
class Mat:
    def __init__(self, mtrx):
        self.update(mtrx)
        self.prec = 3

    def show(self):
        for i, m in enumerate(self.M):
            print("\tRow " + str(i+1) + ": ", self.M[i])
        print("\n")

    def R(self, r):
        if r >= 1 and r <= len(self.M):
            return self.M[r-1]
   
    def augmentI(self):
        L = len(self.M)
        N = []
        for i, m in enumerate(self.M):
            for j, n in enumerate(m):
                m[j] = n * 1.0
            N.extend([m + [0.0 if not(i == ind) else 1.0 for ind in range(L)]])
        self.update(N)

    def update(self, mtrx):
        self.M = mtrx
        
    def addRows(self, rx, tx, con=1):
        N = []
        for i, m in enumerate(self.M):
            if i == rx - 1:
                N.extend([[x+y * con for x, y in zip(self.M[i], self.M[tx - 1])]])
            else:
                N.extend([m])
        self.update(N)

    def subRows(self, rx, tx, con=1):
        N = []
        for i, m in enumerate(self.M):
            if i == rx - 1:
                N.extend([[x - (y * con) for x, y in zip(self.M[i], self.M[tx - 1])]])
            else:
                N.extend([m])
        self.update(N)

    def mulRow(self, rx, con=1):
        N = []
        for i, m in enumerate(self.M):
            if i == rx - 1:
                N.extend([[x * con for x in self.M[i]]])
            else:
                N.extend([m])
        self.update(N)

    def swapRows(self, rX, rY):
        N = []
        for i, m in enumerate(self.M):
            if i == rX - 1:
                N.extend([[x for x in self.M[rY-1]]])
            elif i == rY - 1:
                N.extend([[x for x in self.M[rX-1]]])
            else:
                N.extend([m])
        self.update(N)
        
    def subMatrixRight(self):
        N = []
        for m in self.M:
            N.extend([m[int(len(m)/2):]])
        self.update(N)
    
    def setPrec(self, P=3):
        self.prec = P        
    
    def showT(self):
        formatStr = '{:' + str(6) + '.' + str(self.prec) + 'f}'
        for m in self.M:
            print("| ", end="")
            for n in m:
                print(formatStr.format(n),end=" ")
            print(" |")
        
        
# -----------------------------------------------------------------------------------
        
Mat1 = [[3, 0, 2],
        [2, 0, -2],
        [0, 1, 1]]

M1 = Mat(Mat1)
print("\n")
M1.showT()
print("\n")
print(M1.R(1))
M1.augmentI()
print("-"*60)
M1.showT()
M1.subMatrixRight()
print("\n")
M1.showT()
print("\n")
M1.setPrec(5)
M1.showT()
```
