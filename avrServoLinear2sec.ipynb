```python
# Interactive Python Notebook
# Calculating a table of 8 bit signed values that will be added to a 16 bit pulse width 
# to oscillate a servo motor in a sinusoidal motion.

import numpy as np
import matplotlib.pyplot as plt
from findReplace import subStrings
%matplotlib inline

# pwMax and pwMin determined by bench testing servo motor with signal generator
# desired travTime will be the approximate time the servo takes to move from full
#     left to full right.
# deltaT = 20 mS (Period for requred 50 Hz pulse to servo motor)
# steps is calculated number of steps required to cover travTime
# AVR program must load calculated pwMid to pulse width register and output to servo during setup

pwMax = 2350
pwMin = 450
pwRange = pwMax - pwMin
pwMid = int((pwMax + pwMin) / 2)
travTime = 4
deltaT = .02
steps = int(travTime / deltaT)

print("  Max:", pwMax, "  Min:", pwMin, "  Mid:", pwMid, "  Range:", pwRange, "  Steps:", steps)

# Generate list of values with peaks at pwMax and pwMin
data = []
for t in range(steps):
#    pos = int((np.abs(((t - steps / 4) % steps) - steps / 2) - steps / 4) * pwRange * 2 / steps + pwMid)
    pos = int(pwRange / 2 * np.sin(t * 2 * np.pi / steps) + pwMid)
    data.append(pos)

print("\nPulse Widths:\n", data)
plt.plot(data)
plt.grid(True)
plt.show()
print("\nCalculated Min:", min(data), "  Calculated Max:", max(data))

# Generate list of step sizes that can be added to Pulse Width register
stepSize = []
stepSizeSigned = []
for i, d in enumerate(data):
    size = data[(i + 1)%len(data)] - d
    stepSize.append(size)
    stepSizeSigned.append(size if size >= 0 else 256 + size)

print("\nStep Size\n", stepSize)

plt.plot(stepSize)
plt.grid(True)
plt.show()

# Generating a string containing all 8 bit signed hex values formatted as a table with .db (define byte)
# directives that can be included in the AVR program.

TaddLable = "TABLEaddr" + str(travTime).replace(".","_")
TendLable = "TABLEend" + str(travTime).replace(".","_")

Table = TaddLable + ":\n.db\t"
for i, s in enumerate(stepSizeSigned):
    Table += "$" + hex(s)[2:].zfill(2).upper()
    Table += "\n.db\t" if i % 8 == 7 and i < len(stepSizeSigned) - 1 else\
    "\n" if i >= len(stepSizeSigned) - 1 else ",\t"
Table += TendLable + ":"

print("\nStep Size 8 bit Signed\n", stepSizeSigned)

subs = {"TableAdd":TaddLable, "TableEnd":TendLable, "pwmMid": str(pwMid),
        "pwMax":str(pwMax), "pwMin":str(pwMin)}
        
with open('asm', 'r') as ASM:
    for line in ASM:
        if "TABLE" in line:
            print("\n", Table, "\n")
        else:
            print(subStrings(line, subs), end='')
```
